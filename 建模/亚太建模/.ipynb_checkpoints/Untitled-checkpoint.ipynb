```python
# back:25
def scoreOfBackBend():
    numberList=[32,30,15]
    score=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
        t3,x3,y3,z3=getElement(frame,numberList[2])
        maxCos=-2
        for i in range(len(t1)):
            cos=((x2[i]-x1[i])*(x2[i]-x3[i])+(y2[i]-y1[i])*(y2[i]-y3[i])+(z2[i]-z1[i])*(z2[i]-z3[i]))/(cmath.sqrt((x2[i]-x1[i])**2+(y2[i]-y1[i])**2+(z2[i]-z1[i])**2).real*cmath.sqrt((x2[i]-x3[i])**2+(y2[i]-y3[i])**2+(z2[i]-z3[i])**2).real)
            if maxCos<cos:
                maxCos=cos
        score.append(np.arccos(maxCos)*180/3.14)
    score=np.array(score)
    score=180-score
    score=(100-score)/10
    return score
```

```python
# height of shoulder:25
def scoreOfHeightShoulder():
    numberList=[16,17]
    score=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
        maxHeightlength=0
        for i in range(len(t1)):
            h=abs(z1[i]-z2[i])
            if h>maxHeightlength:
                maxHeightlength=h
        score.append(maxHeightlength)
    score=np.array(score)
    score=(100-score)/10
    return score
```

```python
# arm:25


def arm(numberList):
    peakNumber=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
        sumArea=[]
        for i in range(1,len(t1)):
            
            l1=cmath.sqrt((x1[i]-x1[i-1])**2+(y1[i]-y1[i-1])**2+(z1[i]-z1[i-1])**2)
            l2=cmath.sqrt((x1[i]-x2[i])**2+(y1[i]-y2[i])**2+(z1[i]-z2[i])**2)
            sumArea.append(l1*l2)
        peakValueNumber=0
        for i in range(len(sumArea)):
            if i>0 and i+1<len(sumArea) and sumArea[i].real>sumArea[i-1].real and sumArea[i].real>sumArea[i+1].real:
                peakValueNumber+=1
        peakNumber.append(peakValueNumber)
    peakNumber=np.array(peakNumber)
    return peakNumber
def scoreOfArms():
    numberList1=[18,20]    
    numberList2=[19,21]    
    score1=arm(numberList1)
    score2=arm(numberList2)
    score=score1+score2
    maxt=score.max()
    score-=score.min()
    score=(maxt-score)/maxt*10
    return score
```

```python

# body center:25
def scoreOfBodyCenter():
    numberList=[29,33]
    score=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])

        max1=0
        max2=0
        max3=0
        min1=0
        min2=0
        min3=0
        for i in range(len(t1)):
            t1=cmath.sqrt((x1[i]-x2[i])**2+(y1[i]-y2[i])**2+(z1[i]-z2[i])**2)
            t=(x1[i]-x2[i])/t1
            if max1<t:
                max1=t
            if min1>t:
                min1=t
            t=(y1[i]-y2[i])/t1
            if max2<t:
                max2=t
            if min2>t:
                min2=t
        max1=np.arccos(max1)
        max1*=180/3.14
        max1=90-max1
        max2=np.arccos(max2)
        max2*=180/3.14
        max2=90-max2

        min1=np.arccos(min1)
        min1*=180/3.14
        min1=90-min1
        min2=np.arccos(min2)
        min2*=180/3.14
        min2=90-min2

        max1=max1-min1
        max2=max2-min2
        maxt=max1+max2
        score.append(maxt.real)
    score=np.array(score)
    score/=2.5
    score=10-score
    return score
```

```python
# 行走曲线弯曲程度（路线） :17.5

def straightLine(x,t1,x1,y1,z1):
        a=(y1[len(t1)-1]-y1[0])/(x1[len(t1)-1]-x1[0])
        b=y1[0]-a*x1[0]
        return a*x+b
def scoreOfLineStraight():
    numberList=[25,26]
    score=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
        sumLeft=0
        sumRight=0
        for i in range(len(t1)):
            sumLeft+=y1[i]-straightLine(x1[i],t1,x1,y1,z1)
            sumRight+=y2[i]-straightLine(x2[i],t2,x2,y2,z2)
        score.append((abs(sumLeft+sumRight)/len(t1))/2)
    score=np.array(score)
    score=(70-score)/10
    return score
```

```python
# 7.5
def scoreOfRepeatMotion():
    numberList=[37,36]
    score=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
        
        peakValue=[]
        
        for i in range(len(t1)):
            if i>1 and i+1<len(t1) and z1[i]>z1[i-1] and z1[i]>z1[i+1] and z1[i]>100:
                peakValue.append(z1[i])
        times=0
        for i in range(len(peakValue)):
            if z1.max()-peakValue[i]>=z1.max()*(0.2):
                times+=1
        score.append(times)
    score=np.array(score)
    score=3-score
    return score
# print(scoreOfRepeatMotion())
```

```python
# 25
def scoreOfKneeAngle():
    # right knee angle
    numberListLeft=[11,9,7,27,36]
    # get right middle knee :9,27
    # left knee angle
    numberListRight=[12,10,8,28,37]

    l1=kneeAngle(numberListLeft)
    l2=kneeAngle(numberListRight)
    l1=np.array(l1)
    l2=np.array(l2)
    score=(abs(l1+l2)/2-90)/90*10
    return score


def kneeAngle(numberList):
    kneeAngleAve=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[1])
        t2,x2,y2,z2=getElement(frame,numberList[3])
        y=[]
        x=[]
        z=[]
        for i in range(len(y1)):
            y.append((y1[i]+y2[i])/2)
            x.append((x1[i]+x2[i])/2)
            z.append((z1[i]+z2[i])/2)
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[2])
        a=[]
        b=[]
        c=[]
        for i in range(len(y1)):
            a.append(cmath.sqrt((x1[i]-x2[i])**2+(y1[i]-y2[i])**2+(z1[i]-z2[i])**2))
            b.append(cmath.sqrt((x1[i]-x[i])**2+(y1[i]-y[i])**2+(z1[i]-z[i])**2))    
            c.append(cmath.sqrt((x2[i]-x[i])**2+(y2[i]-y[i])**2+(z2[i]-z[i])**2))
        kneeCos=[]
        for i in range(len(y1)):
            kneeCos.append((b[i]**2+c[i]**2-a[i]**2)/(2*b[i]*c[i]))
        kneenp=np.array(kneeCos)
    #     print(kneenp.max()-kneenp.mean())
        t3,x3,y3,z3=getElement(frame,numberList[4])
        times=[]
        for i in range(len(t3)):
            if z3[i]<=z3.min()+10:
                times.append(i)
        sum=0
        for i in times:
            sum+=kneenp[i]
        sum/=len(times)
        sum=np.arccos(sum)
        kneeAngleAve.append((sum/3.14)*180)
    return kneeAngleAve
```

```python
# 脚步对称的评分:25

def scoreOfStepsSymmetric():
    # set the number1 of body which will be thought
    numberList=[36,37]
    scoreList=[]
    for q in range(len(frameList)):
        frame=frameList[q]
        t1,x1,y1,z1=getElement(frame,numberList[0])
        t2,x2,y2,z2=getElement(frame,numberList[1])
    #   x方向单步长

        stepLeft=[]
        

        for i in range(len(t1)):
            if i>10 and abs(x2[i]-x2[i-10])<3.0 and i+10<len(x2):
                stepLeft.append(x2.iloc[i+5])
                i+=100
        left=[]
        for i in range(len(stepLeft)):
            for j in range(len(stepLeft)):
                if abs(stepLeft[i]-stepLeft[j])>500.0 and abs(stepLeft[i]-stepLeft[j])<1500.0:
                    left.append(abs(stepLeft[i]-stepLeft[j]))

        stepLeft=left
        leftLength=np.array(stepLeft).mean()
        
        
        stepRight=[]
        for i in range(len(x1)):
            if i>10 and abs(x1[i]-x1[i-10])<3.0 and i+10<len(x1):
                stepRight.append(x1.iloc[i+5])
                i+=100
        right=[]
        for i in range(len(stepRight)):
            for j in range(len(stepRight)):
                if abs(stepRight[i]-stepRight[j])>500.0 and abs(stepRight[i]-stepRight[j])<1500.0:
                    right.append(abs(stepRight[i]-stepRight[j]))
        stepRight=right
        rightLength=np.array(stepRight).mean()
        
        
        stepDifferenceValue=abs(leftLength-rightLength)
        score1=(100-stepDifferenceValue)/10
        scoreList.append(score1)
    return np.array(scoreList)
```

```python
# plot a 2d picture
def plotAPicture(x,y,colors,titles):
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    ax.plot(x,y,color=colors)
    ax.set_title(titles)
#     ax.legend()
def plotA3DPicture(x,y,z):
    fig=plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot(x,y,z,label='parametric curve')
def plot2Picture(x1,y1,x2,y2,color1,color2,titles):
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    ax.plot(x1,y1,color=color1)
    ax.plot(x2,y2,color=color2)
    ax.set_title(titles)
```

```python
# 获得对应的老年人的步态数据中的第number个传感器的xyz和time数据
def getElement(frame,number):
    return frame.iloc[:,1],frame.iloc[:,number*3-1],frame.iloc[:,number*3],frame.iloc[:,number*3+1]
```

```python
# 访问文件下所有的路径，读出所有文件

import os, shutil
def readAllFile():
    filenames=eachFile('./data')
    frameList=[]
    for name in filenames:
        frame=pd.read_csv(name)
        frameList.append(frame)
    return frameList
    
    

def eachFile(filepath):
    work_dir = filepath
    for parent, dirnames, filenames in os.walk(work_dir,  followlinks=True):
        fileList=[]
        for filename in filenames:
            file_path = os.path.join(parent, filename)
            fileList.append(file_path)
#             print('文件名：%s' % filename)
#             print('文件完整路径：%s\n' % file_path)
#         print(fileList)
        return fileList

```
